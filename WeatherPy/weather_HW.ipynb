```python
import requests
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
from citipy import citipy
from api_keys import weather_api_key
from pprint import pprint
from scipy.stats import linregress
```

```python
lat = []
lng = []
nearest_city =[]
nearest_ctry =[]
data_dict = []

num_cities = 10

for x in range(num_cities):
    lat_rnd = random.uniform(-90,90)
    lng_rnd = random.uniform(-180,180)
    city = citipy.nearest_city(lat_rnd, lng_rnd)
    city_rnd = city.city_name
    ctry_rnd = city.country_code
    
#     url = "http://api.openweathermap.org/data/2.5/weather?q=London,uk&appid={a7a1f8581953b7105f0c5ef795920a44}"
#     url = "http://api.openweathermap.org/data/2.5/weather?q={city_rnd},{ctry_rnd}&appid={weather_api_key}"
#     url = "http://api.openweathermap.org/data/2.5/weather?lat=lat_rnd&lon=lng_rnd&units=imperial&appid=a7a1f8581953b7105f0c5ef795920a44"
#     url = "http://api.openweathermap.org/data/2.5/weather?lat=' + lat_rnd + '&lon=' + lng_rnd + '&units=imperial&appid=a7a1f8581953b7105f0c5ef795920a44" 
    url = "http://api.openweathermap.org/data/2.5/onecall?lat=" + str(lat_rnd) + "&lon=" + str(lng_rnd) + "&exclude=hourly,current,minutely,alerts&units=imperial&appid=a7a1f8581953b7105f0c5ef795920a44"
    
    request = requests.get(url)
    data = request.json()
    
    lat.append(lat_rnd)
    lng.append(lng_rnd)
    nearest_city.append(city_rnd)
    nearest_ctry.append(ctry_rnd)
    data_dict.append(data)
#     print(f'Nearest city to Lat: {lat[x]} and Long: {lng[x]} is {nearest_city[x]}, {nearest_ctry[x]}')
    pprint(data_dict)
data_dict[0]['daily'][0]['temp']['day']
 
```

```python
temp_day = []
temp_night = []
temp_max = []
temp_min = []
temp_current = []
humidity = []
clouds = []
wind = []

hemisphere = [0 for x in range(num_cities)]

for x in range(num_cities):
#     temp_max.append(data_dict[x]['temp']['max'])
#     print(data_dict[x]['daily'].loc['temp'])
#     temp_min.append(data[x]['daily'][0]['temp']['min'])
    df = pd.DataFrame(data_dict[x]['daily'])
    new_df = df['temp'][x]
    temp_day.append(new_df['day'])
    temp_night.append(new_df['night'])
    humidity.append(df['humidity'][0])
    clouds.append(df['clouds'][0])
    wind.append(df['wind_speed'][0])
    if lat[x] >= 0:
        hemisphere[x] = "Northern"
    else:
        hemisphere[x] = "Southern"
temp_day
temp_night
humidity
clouds
wind
hemisphere
```

```python
    
weather_df = pd.DataFrame(
    {'City': nearest_city,
     'Country':nearest_ctry,
     'Longitude': lng,
     'Latitude': lat,
     'Temperature Day (F)': temp_day,
     'Temperature Night (F)': temp_night,
     'Humidity (%)': humidity,
     'Cloudiness': clouds,
     'Wind Speed (mph)': wind,
     'Hemisphere': hemisphere
    })
weather_df.head
```
